```python
import pandas as pd
from io import StringIO
```

```python
df= pd.read_fwf('task_1_data.log')
```

```python
df.head(10)
```

```python
#刪除前5列-delete the first five rows
df = df.drop(df.index[range(5)])
```

```python
#刪除數據尾部-delete the data at the end of rows
df = df[(df['job starting...'] != '----------------------------------------------') & (df['job starting...'] != 'total point 1000')]
```

```python
#刪除數據中有record successful-delete the data include "record successful"
df = df[(df['job starting...'] != 'record successful')]
```

```python
#delete the data include "result a and b on hold"
df = df[(df['job starting...'] != 'result a and b on hold...')]
```

```python
#delete the data include the word "system"
df = df[df["job starting..."].str.contains("system") == False]
```

```python
#刪除數據中含有finished--delete the data include "finished"
df = df[df["job starting..."].str.contains("finished") == False]
```

```python
#we get the cleanding data 
df.head(10)
```

```python
#let the row1 and row2 combine to a column
df = pd.DataFrame(df['job starting...'].values.reshape(-1,2),
                   columns=['x', 'a'])
```

```python
df.head(10)
```

```python
#split the data to three column
df[['a', 'b']] = df['a'].str.split(',', expand=True)
```

```python
df.head(10)
```

```python
#split the column to 5 columns
df[['z','x', 'y']] = df['x'].str.split(',', expand=True)
```

```python
df.head(10)
```

```python
#split the column to 6 columns
df[['p', 'x']] = df['x'].str.split('(', expand=True)
```

```python
df.head(10)
```

```python
#split the column to 7 columns
df[['y', 'q']] = df['y'].str.split(')', expand=True)
```

```python
df.head(10)
```

```python
#split the column to 8 columns
df[['r', 'a']] = df['a'].str.split('=', expand=True)
```

```python
df.head(10)
```

```python
#split the column to 9 columns
df[['s', 'b']] = df['b'].str.split('=', expand=True)
```

```python
df.head(10)
```

```python
#delete the columns that we do not need
df.drop(columns=['b','z','p','q','r', 's'])
```

```python
#reorder the columns
df=df[['x','y','a']]
```

```python
df.head(10)
```

```python
#output csv file and name the file "file.csv"
df=df.to_csv('file.csv')
```

```python
#collecting all the value in the csv file
ALL= 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
#build the result analyze a ouput
dropdown_result_a=widgets.Dropdown(options=unique_sorted_values_plus_ALL(df.a))

output_result_a=widgets.Output()

def dropdown_result_a_eventhandler(change):
    output_result_a.clear_output()
    with output_result_a:
        if (change.new==ALL):
            display(df)
        else:
            display(df[df.a==change.new])
dropdown_result_a.observe(dropdown_result_a_eventhandler, names='value')

display(dropdown_result_a)
```

```python
#build the result analyze b ouput
dropdown_result_b=widgets.Dropdown(options=unique_sorted_values_plus_ALL(df.b))

output_result_b=widgets.Output()

def dropdown_result_b_eventhandler(change):
    output_result_b.clear_output()
    with output_result_b:
        if (change.new==ALL):
            display(df)
        else:
            display(df[df.b==change.new])
dropdown_result_b.observe(dropdown_result_b_eventhandler, names='value')

display(dropdown_result_b)
```
